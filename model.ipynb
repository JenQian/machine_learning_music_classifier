```python
categoricals = [
    'acousticness',
    'danceability',
    'duration_ms energy',
    'instrumentalness',
    'key_from_api',
    'liveness',
    'loudness',
    'mode_from_api',
    'speechiness',
    'tempo',
    'time_signature_from_api',
    'valence',
    'explicit',
    'release_year'
]
```

```python
X = spotify_df[[
    'acousticness',
    'danceability',
    'duration_ms',
    'energy',
    'instrumentalness',
    'key_from_api',
    'liveness',
    'loudness',
    'mode_from_api',
    'speechiness',
    'tempo',
    'time_signature_from_api',
    'valence',
    'explicit',
    'release_year'
]]
X
```

```python
y= spotify_df['genre_coded']
y
```

```python

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y)

x_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier



rf = RandomForestClassifier()

params = {
    'n_estimators': [235],
    'max_depth': [13],
    #'min_samples_split': [13],
    #'min_samples_leaf': [9,7]
}

clf = GridSearchCV(rf, params, verbose=10, n_jobs=-1)

#model = clf.fit(x_train, y_train)
model = clf.fit(X, y)
```

```python
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
model.score(x_test, y_test)
```

```python
print(f'best score: {clf.best_score_}')
print(f'best score: {clf.best_params_}')
```
